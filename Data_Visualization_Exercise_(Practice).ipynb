#Data visualization:

Objective: Biggest Tomato Producers in 1997

Original visualization:
image.png

```python
df_new = df[['Area','Element','Item','Value','Year']]
df_new.head()
```

```python
df_filter1 = df_new['Element']=='Production'
tableau1= df_new[df_filter1]
tableau1.head()
```

```python
df_filter_tomatoes = tableau1['Item']=='Tomatoes'
tableau_tomatoes=tableau1[df_filter_tomatoes]
tableau_tomatoes.head()
```

```python
filter_year = tableau_tomatoes['Year']==1997
tomatoes_1997 = tableau_tomatoes[filter_year]
tomatoes_1997.head()
```

```python
tomatoes_1997.groupby("Area")['Value'].max().sort_values(ascending=False).head(10)
```

```python
df_x=tomatoes_1997.nlargest(13, 'Value')
df_x
```

```python
df_y=df_x.drop([9930])
df_y
```

```python
ax=sns.barplot(data=df_y, y='Area',x='Value');
plt.style.use('seaborn-v0_8-poster')
plt.xlabel('Tomatoes production in 1997 (Tonnes)')
plt.ylabel('Country')
plt.title('Biggest Tomato producers in 1997',size=28);
ax.bar_label(ax.containers[0],fmt='%.0f');
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
